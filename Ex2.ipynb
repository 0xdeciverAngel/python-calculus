```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01
r = 10
# Need one more for the initial values
t = np.arange(0, 4*np.pi+dt, dt)
x = r * np.cos(t)
y = r * np.sin(t) 
z = 0.1*t

# Plot
fig = fig = plt.figure (figsize=[8,4.5], facecolor = 'red')
ax = fig.gca(projection='3d')
ax.set_facecolor ( (0.5, 1., 0.) )

ax.plot(x, y, z , 'b-', lw=2.0)
ax.set_xlabel ("X Axis")
ax.set_ylabel ("Y Axis")
ax.set_zlabel ("Z Axis")
ax.set_title ("Helix")
ax.set_xticks ( np.arange (-r, r, r/2) )
ax.set_yticks ( np.arange (-r, r, r/2) )
ax.set_zticks ( [0, 0.5, 1.0] )
ax.view_init (30, 60)
plt.show()
```
